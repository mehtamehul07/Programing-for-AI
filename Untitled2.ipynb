```python
# Import Librabries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans, DBSCAN
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
import numpy as np

# Load datasets

iris = load_iris()
X= iris.data[:, :2]  #Take only first 2 features for visualization

#===============================================================================

# Data Preprocessing (Importtant for DBSCAN)

#===============================================================================

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#===============================================================================

# K-MEANS CLUSTRING

#===============================================================================

kmeans= KMeans(n_clusters=3, random_state=0)
kmeans_labels = kmeans.fit_predict(X_scaled)
print("Labels:", kmeans_labels)

print("K-Means Labels:\n", kmeans_labels)

# Plot K-Means
plt.figure()
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans_labels)
plt.title("K-Means Clustring (Iris Datasets)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

#===============================================================================

# DBSCAN CLUSTRING

#===============================================================================

dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)

print("DBSCAN Labels:\n", dbscan_labels)

# Plot DBSCAN
plt.figure()
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=dbscan_labels)
plt.title("DBSCAN Clustring (Iris Datasets)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()


# Step 1 : Choose min-samples
min_samples = 4

# Step 2 : Compute nearest neighbors
neighbors = NearestNeighbors(n_neighbors=min_samples)
neighbors_fit = neighbors.fit(X_scaled)

# Step 3 : Get distance
distances, indices = neighbors_fit.kneighbors(X_scaled)
print("Distances:\n", distances)
print("Indices:\n", indices)


# Step 4 : Sort Distances
distances = np.sort(distances[:, -1])

# Step 5 : Plot
plt.plot(distances)
plt.xlabel("Points sorted by distances")
plt.ylabel("Distance")
plt.title("K-Distance Graph")
plt.show()
```
