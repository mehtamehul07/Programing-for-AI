```python
#series


import pandas as pd
marks = pd.Series([78,85,90,66])
print(marks)


print("Type of values: ", marks.values)
print("Type of index: ", marks.index)

```

```python
# Series with explicit index



import pandas as pd
marks = pd.Series([78,85,90,66], index = (['a', 'b', 'c', 'd']))
print(marks)


print("Index: ", marks.index)

print("Access Element: ", marks["a"])
```

```python
import pandas as pd
marks = pd.Series([78,85,90,66], index = (['1', '2', '3', '4']))
print(marks[1])
print (marks['1'])
print (marks.iloc[3])
print (marks.loc['3'])

print (marks.iloc[1:2]) #implicit
print (marks.loc['1':'2']) #explicit
```

```python
# create series with dictionary

import pandas as pd
dict={"A":10, "B":20, "C":30}
marks=pd.Series(dict)
print("Series Object From Dict\n", marks)
print("Acces with label-based slicing(exp)\n", marks["A":"C"])
print("Acces with position-based slicing(imp)\n", marks[0:2])

print("Acces with label-based slicing(exp)\n", marks.iloc[0:2])
print("Acces with position-based slicing(imp)\n", marks.loc['0':'2'])

```
