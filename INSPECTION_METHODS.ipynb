```python
# Data Inspection Methods

print("INSPECTION METHODS PYTHON (PANDA)")


import pandas as pd
data ={
    "Name":["Amit", "Riya", "Karan"],
    "Marks":[70,85,90]
}
df=pd.DataFrame(data)

print("DATA: ")
print(df)
print("\n")
print("SHAPE: ")
print(df.shape)
print("\n")
print("DATA TYPE: ")
print(df.dtypes)
print("\n")
print("COLUMN: ")
print(df.columns)
print("\n")
print("INDEX: ")
print(df.index)
print("\n")
print("HEAD: ")
print(df.head())
print("\n")
print("TAIL:")
print(df.tail())
print("\n")
print("INFO: ")
print(df.info())
print("\n")
print("DESCRIBE: ")
print(df.describe())
print("\n")

print("CSV FILE: ")
db=pd.read_csv("Inspection.csv")
print(db)
db

print("ISNULL: ")
print(db.isnull())
print("\n")
print("ISNULL.SUM: ")
print(db.isnull().sum())
print("\n")
print("COUNT OF GRADE: ")
db['Grade'].value_counts(dropna=False)
print("\n")
print(db.notnull)
```
