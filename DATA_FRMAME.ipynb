```python
# DATA FRAME

import pandas as pd
data ={"Name":["Amit", "Riya", "Karan"],"Marks":[70,85,90]}
df=pd.DataFrame(data)
print(df)
df
```

```python
# Create Data frame from list of dictionary

import pandas as pd
data=[{"Name":"Amit", "Marks":70},{"Name":"Riya", "Marks":85},{"Name":"Karan", "Marks":90}]
df=pd.DataFrame(data)
print(df)
df
```

```python
# Create data frame from list of lists

import pandas as pd
data=[["Amit", 70], ["Riya", 85], ["Karan", 90]]
df=pd.DataFrame(data, columns=["Name", "Marks"])
print(df)
df
```

```python
#  Create Data frame from dictionary of series

import pandas as pd
s1=pd.Series([70,85,90])
s2=pd.Series(["Amit", "Riya", "Karan"])
df=pd.DataFrame({"Name":s2, "Marks":s1})
print(df)
df
```
